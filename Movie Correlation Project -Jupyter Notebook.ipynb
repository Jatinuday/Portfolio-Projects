```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create 

#Read in the data
df = pd.read_csv(r'C:\Users\jatin uday\Downloads\movies.csv')
```

```python
#  Lets look at the data

df.head()
```

```python
#Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%' .format(col, pct_missing))
```

```python
#Data types for our columns

df.dtypes
```

```python
#Change data type of columns
df['budget'] = df['budget'].astype('float64')
df['gross'] = df['gross'].astype('float64')
```

```python
df
```

```python
# Handle missing or invalid values in the 'released' column
df['released'] = df['released'].fillna('0000').astype(str, errors='ignore')

# Fill missing years with a default year (e.g., 0000)
df['released'] = df['released'].replace('', '0000', regex=True)

# Create correct Year column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

# Drop rows with missing values
df = df.dropna()
df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatterplot with budget vs gross revenue


import matplotlib.pyplot as plt

# Create a scatterplot of budget vs gross revenue
plt.scatter(x=df['budget'], y=df['gross'])

# Add labels and title to the plot
plt.xlabel('Budget (in millions)')
plt.ylabel('Gross Revenue (in millions)')
plt.title('Scatterplot of Budget vs Gross Revenue')

# Display the plot
plt.show()
```

```python
# Look for correlation
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numeric_cols].corr()
print(correlation_matrix)
```

```python
#High correlation between year and score

correlation_matrix = df[numeric_cols].corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matirx for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
